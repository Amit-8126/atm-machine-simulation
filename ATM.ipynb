```python
from accounts import accounts

def ATM():
    while True:
        
        user=input("what you want to do?'widrawal','deposit','balance','exit' or history:").lower()
        if user=="history":
            u=input("please enter your id:")
            if accounts[u]["history"]==[]:
                print("there is no history")
            else:
                print(accounts[u]["history"])
        elif user=="exit":
            print("ATM Closed.Thank you!")
            break
        elif user=="balance":
             u=input("please enter your id:")
             if u in accounts:
                 print(f"your balance is {accounts[u]['balance']}")
             else:
                 print("no user found")
    
        elif user=="widrawal":
            u=input("please enter your id:")
            if u in accounts:
                wid=int(input("how much money you want to widraw?:"))
                if wid>accounts[u]["balance"]:
                    print(f"insufficient funds.Your current balance is {accounts[u]['balance']}")
                elif wid<=accounts[u]["balance"]:
                    accounts[u]["balance"]-=wid
                    accounts[u]["history"].append({
                        "type":"widrawal",
                        "amount":wid,
                        "balance after":accounts[u]["balance"]
                    }
                        
                )
                    print(f"success.your current balance is {accounts[u]['balance']}")
        
            else:
                 print("no user found")
    
        elif user=="deposit":
            u=input("please enter your id:")
            if u in accounts:
                wid=int(input("how much money you want to deposit?:"))
                accounts[u]["balance"]+=wid
                accounts[u]["history"].append({
                        "type":"deposit",
                        "amount":wid,
                        "balance after":accounts[u]["balance"]
                }
             )
                print(f"success.your current balance is {accounts[u]['balance']}") 
            else:
                 print("no user found")
         
        else:
            print("invalid operation!Try again.")
        

    
       
    
```
